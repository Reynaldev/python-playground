```python
from sklearn.preprocessing import LabelEncoder

# Sample categorical data
fruits = ["apple", "banana", "orange", "apple", "orange", "banana", "apple"]

# Step 1: Initialize LabelEncoder
encoder = LabelEncoder()

# Step 2: Fit and transform the categorical values into numeric labels
encoded = encoder.fit_transform(fruits)
print("Encoded labels:", encoded)

# Step 3: Inverse transform back to original categories
decoded = encoder.inverse_transform(encoded)
print("Decoded labels:", decoded)
```
